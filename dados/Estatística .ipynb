# Medida de Tendência Central  

```python
import pandas as pd
import numpy as np


df = pd.DataFrame(data={
    'Fulano': [7.5, 8.0, 7.0, 8.0, 8.0, 8.5, 7.0],
    'Beltrano': [8, 10, 4.0, 8.0, 6.0, 10, 8.0],
    'Sicrano': [10, 2.0, 0.5, 1.0, 3.0, 9.5, 10]
}, index=['Matemática', 'Português', 'Inglês', 'Geografia', 'História', 'Física', 'Química'])


df.reset_index(inplace=True)
df.rename(columns={'index': 'Matéria'}, inplace=True)
```

```python
df
```

```python
#média
df[['Fulano', 'Beltrano', 'Sicrano']].mean()
```

```python
#mediana
df[['Fulano', 'Beltrano', 'Sicrano']].median()
```

```python
#moda
df[['Fulano', 'Beltrano', 'Sicrano']].mode().T
```

Nesse Resultado vemos Fulano e Beltrano tendo a mesma média.

O nosso intuito nessa exemplo é mostra o que o mean(), acaba não sendo eficaz em caso diferentes.

Teremos que fazer uma nova média pra verificar se os dados da média pode ser diferente com a **Média Aritmética**.

# Desvio Médio Absoluto


```python
df['Desvio'] = df.Beltrano            #criando nova coluna com os dados do Beltrano
media_abs = df.Beltrano.mean()        #criando uma variavel para media de beltrano
df['Desvio'] = df.Desvio - media_abs  #Aqui é subtração da coluna Desvio com a media de Beltrano
df['Desvios'] = df.Desvio.abs()      #Refazendo a coluna com os trocas de sinais
df
```

### **O uso do abs(),** é mais ultilizados para fazer a troca de sinais de numeros que foram colocados errados em uma tabela, ele faz a conversão dos sinais matendo sempre positivos e não interfer nos numeros.

O MAD (Mean Absolute Deviation) é uma medida de dispersão que calcula a média das diferenças absolutas entre os valores e a média (ou mediana) dos dados. Ele é útil para entender a variabilidade dos dados de forma robusta, sendo menos sensível a valores extremos (outliers) em comparação com o desvio padrão.

Quando usar:
* Quando você deseja medir a dispersão sem ser influenciado por outliers.
* Em dados com distribuições assimétricas ou com valores extremos.

```python
df.Desvios.mean() #Desvio médio absoluto
```
